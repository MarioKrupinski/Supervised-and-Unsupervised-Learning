<h1> Projekt Zusammenfassung

Autoren:
Michelle Mai, Antionio Piloro; Mario Krupinski

<h2> Motivation: </h2>
Emotionserkennung anhand visueller Merkmale spielt in einer Vielzahl moderner Anwendungen eine wachsende Rolle. Sie ermöglicht es Systemen, menschliche Zustände besser zu interpretieren und auf natürliche Weise zu reagieren – sei es im Bereich der Mensch-Maschine-Interaktion, im Gesundheitssektor, in Lernsystemen oder bei Sicherheitsanwendungen. Gleichzeitig ist das automatische Erkennen von Emotionen technisch herausfordernd, da Gesichtsausdrücke stark variieren und oft subtil ausfallen.

Durch die Kombination zweier umfangreicher Datensätze (FER-2013 und AffectNet) entsteht ein besonders großes und vielfältiges Trainingsset, das eine robuste Modellierung erlaubt. Diese Datenbasis ist ideal, um moderne Deep-Learning-Methoden auf ein anspruchsvolles, praxisrelevantes Problem anzuwenden und daraus wertvolle Erkenntnisse für Emotionserkennungssysteme abzuleiten.

<h2> Ziel </h2>

Ziel des Projekts ist es, ein leistungsfähiges Modell zur Klassifikation von sieben grundlegenden Emotionen aus Gesichtsbildern zu entwickeln: angry, disgust, fear, happy, neutral, sad und surprise.


<h2> Vorgehen </h2>

Im ersten Schritt werden die beiden Datansätze FER-2013 und AffectNet analysiert, kombiniert, bereinigt, augmentiert und für das Training vorbereitet. Anschließend wird damit ein Proof-of-Concept CNN trainiert, dessen Accuracy beweist, dass sich die Daten für eine Klassifizierung eignen. Siehe *Notebook 02_Preprocessing*

Danach werden verschiedene Modelle implementiert, optimiert und evaluiert. Diese werden anschließend miteinander verglichen.
Am Ende beschreiben wir unsere Learning saus dem Projekt.
